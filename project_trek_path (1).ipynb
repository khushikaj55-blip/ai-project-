```python
import pandas as pd
```

```python
df = pd.read_csv("/trek_data - trek_data.csv")   # use your exact file name
df.head()
```

```python
df.info()
```

```python

easy_df = df[df["Difficulty"] == "Easy"]

easy_df.head()
```

```python
import matplotlib.pyplot as plt

# Top 5 locations with most easy trails
top_locations = easy_df["Location"].value_counts().head(5)

top_locations.plot(kind="bar")
plt.title("Top 5 Locations with Easy Trails")
plt.xlabel("Location")
plt.ylabel("Number of Trails")
plt.xticks(rotation=30)
plt.show()
```

```python
moderate_df = df[df["Difficulty"] == "Moderate"]

moderate_df.head()
```

```python
moderate_location = moderate_df["Location"].value_counts().head(5)
print(moderate_location)
```

```python
import matplotlib.pyplot as plt

moderate_location.sort_values().plot(kind="barh")
plt.title("Top Locations for Moderate Trails")
plt.xlabel("Number of Trails")
plt.ylabel("Location")
plt.tight_layout()
plt.show()
```

```python
hard_df = df[df["Difficulty"] == "Hard"]

hard_df.head()
```

```python
hard_location = hard_df["Location"].value_counts().head(5)
print(hard_location)
```

```python
import matplotlib.pyplot as plt

hard_location.sort_values().plot(kind="barh")
plt.title("Top Locations for Hard Trails")
plt.xlabel("Number of Trails")
plt.ylabel("Location")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count each difficulty type
difficulty_counts = df["Difficulty"].value_counts()

# Plot
difficulty_counts.plot(kind="bar")

plt.title("Distribution of Trails by Difficulty")
plt.xlabel("Difficulty")
plt.ylabel("Number of Trails")

plt.xticks(rotation=0)
plt.show()
```

```python
# Group by location and calculate average rating + total reviews
location_stats = df.groupby("Location").agg({
    "Average_rating": "mean",
    "number_of_reviews": "sum"
})

# Sort by rating (best first)
location_stats = location_stats.sort_values(by="Average_rating", ascending=False)

# Show top 5
print(location_stats.head(5))
```

```python
import matplotlib.pyplot as plt

top_locations = location_stats.head(5)

top_locations["Average_rating"].plot(kind="barh")

plt.title("Top Locations by Average Rating")
plt.xlabel("Average Rating")
plt.ylabel("Location")
plt.show()
```

```python
# Filter high rated trails
high_rating_df = df[df["Average_rating"] >= 4.5]

# Show top 10
high_rating_df = high_rating_df.sort_values(by="Average_rating", ascending=False)

print(high_rating_df[["Trail_name", "Location", "Difficulty", "Average_rating"]].head(10))
```

```python
import matplotlib.pyplot as plt

high = len(df[df["Average_rating"] >= 4.5])
low = len(df[df["Average_rating"] < 4.5])

plt.bar(["High Rating", "Other"], [high, low])
plt.title("High Rated vs Other Trails")
plt.ylabel("Number of Trails")
plt.show()
```

```python
# Filter low rated trails
low_rating_df = df[df["Average_rating"] < 4.5]

# Sort lowest first
low_rating_df = low_rating_df.sort_values(by="Average_rating", ascending=True)

print(low_rating_df[["Trail_name", "Location", "Difficulty", "Average_rating"]].head(10))
```

```python
# Filter low rated trails
low_rating_df = df[df["Average_rating"] < 4.5]

# Sort lowest first
low_rating_df = low_rating_df.sort_values(by="Average_rating", ascending=True)

print(low_rating_df[["Trail_name", "Location", "Difficulty", "Average_rating"]].head(10))
```

```python
import matplotlib.pyplot as plt

low = len(df[df["Average_rating"] < 4.5])
high = len(df[df["Average_rating"] >= 4.5])

plt.bar(["Low Rating", "High Rating"], [low, high])
plt.title("Low vs High Rated Trails")
plt.ylabel("Number of Trails")
plt.show()
```
